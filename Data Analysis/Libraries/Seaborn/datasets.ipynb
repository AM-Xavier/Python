```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='YlGnBu')
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='RdBu')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='sex', palette='RdYlBu')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', palette="YlGnBu")
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', dodge=True, palette="YlGnBu")
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='kde', fill=True, cmap='RdYlBu')
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='hex', cmap='RdYlBu')
```
